### 2. Carilah data outlier dan beri penjelasan (insight) pada feature/column Distance ! Tambahkan plot untuk menampilkan adanya data outlier!

```python
def find_anomalies(data):
    anomalies = []
    q1 = data.quantile(0.25)
    q3 = data.quantile(0.75)
    iqr = q3 - q1
    limit = iqr*1.5
    print(f'IQR: {iqr}')

    limit_bawah = q1 - limit
    limit_atas = q3 + limit
    print(f'limit_bawah: {limit_bawah}')
    print(f'limit_atas: {limit_atas}')

    for outlier in data:
        if outlier < limit_bawah or outlier > limit_atas:
            anomalies.append(outlier)
            counter = (len(anomalies))
    print(f'total anomalies: {counter}')
    print(f'anomalies: {anomalies}')        
    return anomalies
```

```python
find_anomalies(df['Distance'])
```

```python
sns.boxplot(df['Distance'])
plt.title('Boxplot dari Distance')
print('Berdasarkan perhitungan dan grafik di atas, ada 204 property yang merupakan outlier. Outlier dalam feature ini memiliki makna bahwa ke 204 property tersebut memiliki jarak yang cukup jauh dari CBD (Central Business District). Insight untuk konsumen: tidak disarankan untuk membeli property yang merupakan outlier karena jaraknya yang jauh dari CBD')
plt.savefig('3.png')
```

### 7. Lakukan Uji Hipotesis untuk menguji apakah ada perbedaan harga properti antar Region Name! (Anggap Price memiliki distribusi normal)

```python
stats.f_oneway(df[df['Regionname']=='Southern Metropolitan']['Price'], 
df[df['Regionname']=='Eastern Metropolitan']['Price'], 
df[df['Regionname']=='Northern Metropolitan']['Price'],
df[df['Regionname']=='Western Metropolitan']['Price'],
df[df['Regionname']=='South-Eastern Metropolitan']['Price'])
# berdasarkan perhitungan di atas ditemukan bahwa p value < alpha (0.05). Artinya adalah setidaknya ada 2 grup yang berbeda secara signifikan, tetapi kita tidak bisa menentukan grup mana yang berbeda.
```
